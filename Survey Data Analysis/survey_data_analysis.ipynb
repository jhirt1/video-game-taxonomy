## Categories

### Recent Categories Assigned 

```python
recent_categories_df = response_df[recent_categories].copy()
recent_categories_df.head(3)
```

```python
recent_count = {'NAN':0}

for idx, row in recent_categories_df.iterrows():
    for col in recent_categories:
        if pd.isna(row[col]):
            recent_count['NAN'] += 1
        else:
            if row[col] not in recent_count:
                recent_count[row[col]] = 1
            else:
                recent_count[row[col]] += 1
```

```python
recent_count
```

```python
# Convert recent_count dictionary to a markdown table
recent_count_table = "| Category         | Count |\n|------------------|-------|\n"
for category, count in recent_count.items():
    recent_count_table += f"| {category:<16} | {count:<5} |\n"

print(recent_count_table)
```
